# **Modelos**

Na primeira fase da modelagem preditiva, foi empregada uma abordagem estatística baseada em séries temporais, utilizando o modelo ARIMA (AutoRegressive Integrated Moving Average). Este modelo é reconhecido por sua capacidade de capturar padrões históricos em séries temporais univariadas, o que é particularmente relevante para a previsão de indicadores educacionais ao longo do tempo.

Iniciamos com um modelo univariado focando na variável Ratio, que representa a razão de conclusão em relação ao ingresso no ensino superior. O conjunto de dados foi dividido em dois subconjuntos: 80% dos dados foram utilizados para treinamento do modelo, enquanto os 20% restantes compuseram o conjunto de teste, destinado à validação das previsões.

O modelo ARIMA foi ajustado com os parâmetros (2, 2, 1), onde:

  p = 2 indica a ordem do componente autorregressivo,

  d = 2 refere-se ao número de diferenciações necessárias para tornar a série estacionária,

  q = 1 representa a ordem do componente de média móvel.


**Validação**

A avaliação do desempenho do modelo foi realizada por meio do Root Mean Squared Error (RMSE), uma métrica amplamente utilizada para medir a magnitude média dos erros de previsão. O valor de RMSE obtido foi de 0,0244, indicando um erro médio relativamente baixo entre as previsões e os valores reais observados.

A comparação visual entre os dados reais e os valores previstos foi realizada por meio de um gráfico de linha, no qual as previsões mostraram boa aderência às tendências observadas nos dados de teste. Adicionalmente, o intervalo de confiança das previsões foi plotado, fornecendo uma medida de incerteza associada às estimativas.

Esses resultados iniciais demonstram a viabilidade do modelo ARIMA na previsão da taxa de conclusão no ensino superior. Para as próximas etapas, estão previstos o refinamento dos parâmetros do modelo e a exploração de abordagens complementares, com o objetivo de aumentar ainda mais a acurácia e a robustez das previsões.

```python
from statsmodels.tsa.arima.model import ARIMA
from sklearn.metrics import mean_squared_error

# Divida os dados em treino e teste.
train_size = int(len(df) * 0.8)
train, test = df[0:train_size], df[train_size:len(df)]

# Ajuste o modelo ARIMA no conjunto de dados de treino.
model_train = ARIMA(train['Ratio'], order=(2, 2, 1))
model_train_fit = model_train.fit()

# Faça a previsão no conjunto de dados de teste.
test_forecast = model_train_fit.get_forecast(steps=len(test))
test_forecast_series = pd.Series(test_forecast.predicted_mean, index=test.index)

# Calcule o erro quadrático médio.
mse = mean_squared_error(test['Ratio'], test_forecast_series)
rmse = mse**0.5

# Crie um gráfico para comparar a previsão com os dados reais de teste.
plt.figure(figsize=(14,7))
plt.plot(train['NU_ANO_CENSO'],train['Ratio'], label='Training Data')
plt.plot(test['NU_ANO_CENSO'],test['Ratio'], label='Actual Data', color='orange')
plt.plot(test['NU_ANO_CENSO'],test_forecast_series, label='Forecasted Data', color='green')

plt.fill_between(test['NU_ANO_CENSO'],
                 test_forecast.conf_int().iloc[:, 0],
                 test_forecast.conf_int().iloc[:, 1],
                 color='k', alpha=.15)
plt.title('ARIMA Model Evaluation')
plt.xlabel('Year')
plt.ylabel('Ingress Conclusion Ratio')
plt.legend()
plt.show()

print('RMSE:', rmse)
```

Na segunda etapa de modelagem, foi testado o modelo Prophet



```python
#Importar bibliotecas
import sqlite3
import os
import pandas as pd
from prophet import Prophet
import matplotlib.pyplot as plt
from sklearn.metrics import mean_absolute_error, mean_squared_error

from google.colab import drive
drive.mount('/content/drive')

os.chdir('/content/drive/MyDrive/mack')

# Conectar ao banco
conn = sqlite3.connect("inep.db")
cursor = conn.cursor()

# Carregar dados
df2 = pd.read_sql(
    """
    SELECT NU_ANO_CENSO AS ano,
       SUM(QT_ING) AS ingressantes,
       SUM(QT_CONC) AS concluintes
FROM microdados
GROUP BY NU_ANO_CENSO
ORDER BY NU_ANO_CENSO
    """, con=conn
    )

# Calcular evasão
df2['evasao'] = df2['ingressantes'] - df2['concluintes']

# Preparar para Prophet
df2_prophet = df2[['ano', 'evasao']].rename(columns={'ano': 'ds', 'evasao': 'y'})
df2_prophet['ds'] = pd.to_datetime(df2_prophet['ds'], format='%Y')

# Instanciar e ajustar o modelo
modelo = Prophet(yearly_seasonality=False)
modelo.fit(df2_prophet)

# Criar futuro
futuro = modelo.make_future_dataframe(periods=3, freq='YE')
previsao = modelo.predict(futuro)

# Visualizar
modelo.plot(previsao)
plt.title('Projeção de evasão no ensino superior até 2026')
plt.xlabel('Ano')
plt.ylabel('Evasão (ingressantes - concluintes)')
plt.show()
```

```python
import pandas as pd
import numpy as np
from prophet import Prophet
from sklearn.metrics import mean_absolute_error, mean_squared_error

# Preparar os dados
df2['ds'] = pd.to_datetime(df2['ano'].astype(str) + '-12-31')
df2['y'] = df2['evasao']
df2_modelo = df2[['ds', 'y']].copy()

# Separar treino e teste
df2_treino = df2_modelo[df2_modelo['ds'] < '2021-01-01']
df2_teste = df2_modelo[df2_modelo['ds'] >= '2021-01-01']

# Treinar o modelo Prophet
modelo = Prophet(yearly_seasonality=False)
modelo.fit(df2_treino)

# Criar datas futuras
futuro = modelo.make_future_dataframe(periods=len(df2_teste), freq='Y')
previsao = modelo.predict(futuro)

# Combinar previsões com valores reais
df2_resultado = pd.merge(df2_teste, previsao[['ds', 'yhat']], on='ds', how='inner')

# Avaliação
y_true = df2_resultado['y'].values
y_pred = df2_resultado['yhat'].values

mae = mean_absolute_error(y_true, y_pred)
rmse = np.sqrt(mean_squared_error(y_true, y_pred))
mape = np.mean(np.abs((y_true - y_pred) / y_true)) * 100

print(f"MAE:  {mae:.2f}")
print(f"RMSE: {rmse:.2f}")
print(f"MAPE: {mape:.2f}%")
```
