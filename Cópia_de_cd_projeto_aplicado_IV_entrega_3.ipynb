# **Introdução**

A qualidade da educação é um dos Objetivos de Desenvolvimento Sustentável (ODS) da Organização das Nações Unidas (ONU), especificamente o tema 4, que visa assegurar a educação inclusiva, equitativa e de qualidade, promovendo oportunidades de aprendizagem ao longo da vida para todos. Este projeto de insere-se no contexto da análise da taxa de conclusão no ensino superior.

Será desenvolvido um modelo de séries temporais com o intuito de analisar o comportamento dos alunos matriculados no ensino superior, verificando o percentual de alunos que concluem e o percentual que não conclui o curso ao longo dos anos analisados. O problema selecionado aborda a identificação das principais causas que influenciam a não conclusão dos cursos e a predição da probabilidade de alunos com determinadas características não concluírem o ensino superior.

```python
df = pd.read_sql(
    """
    SELECT
        NU_ANO_CENSO
        , COUNT(DISTINCT CO_MUNICIPIO) AS qtMunicipios
        , COUNT(DISTINCT CO_CURSO) AS qtCursos
        , COUNT(DISTINCT CO_IES) AS qtInstituicoes
        , SUM(QT_ING) AS qtIngressantes
        , SUM(QT_CONC) AS qtConcluintes
    FROM microdados
    GROUP BY NU_ANO_CENSO
    """, con=conn
    )
```

```python
df
```

```python
df_melted = df.melt(id_vars=["NU_ANO_CENSO"],
                     value_vars=["qtIngressantes", "qtConcluintes"],
                     var_name="Categoria", value_name="Quantidade")

# Set Seaborn theme
sns.set_style("whitegrid")

# Create the plot
plt.figure(figsize=(10, 5))
sns.lineplot(data=df_melted, x="NU_ANO_CENSO", y="Quantidade", hue="Categoria", marker="o")

# Customize the plot
plt.xlabel("Ano do Censo")
plt.ylabel("Quantidade")
plt.title("Evolução de Ingressantes e Concluintes por Ano")
plt.legend(title="Categoria")

# Show the plot
plt.show()
```

```python
df["Ratio"] = df["qtConcluintes"] / df["qtIngressantes"]

plt.figure(figsize=(10, 5))
sns.lineplot(data=df, x="NU_ANO_CENSO", y="Ratio", marker="o")
plt.xlabel("Ano do Censo")
plt.ylabel("Taxa de Conclusão")
plt.title("Evolução da Taxa de Conclusão (Concluintes / Ingressantes)")
plt.grid(True)
plt.show()
```

No gráfico acima podemos ver o efeito da COVID na taxa de concluintes, apesar de que ela já vinha diminuindo desde antes da COVID

```python
df_melted = df.melt(id_vars=["NU_ANO_CENSO"],
                     value_vars=["qtMunicipios", "qtCursos", "qtInstituicoes"],
                     var_name="Categoria", value_name="Quantidade")

plt.figure(figsize=(10, 5))
sns.lineplot(data=df_melted, x="NU_ANO_CENSO", y="Quantidade", hue="Categoria", marker="o")
plt.xlabel("Ano do Censo")
plt.ylabel("Quantidade")
plt.title("Crescimento de Instituições, Cursos e Municípios")
plt.legend(title="Categoria")
plt.grid(True)
plt.show()
```

```python
df["growthIngressantes"] = df["qtIngressantes"].pct_change() * 100
df["growthConcluintes"] = df["qtConcluintes"].pct_change() * 100

df_melted = df.melt(id_vars=["NU_ANO_CENSO"],
                     value_vars=["growthIngressantes", "growthConcluintes"],
                     var_name="Categoria", value_name="Crescimento (%)")

plt.figure(figsize=(10, 5))
sns.barplot(data=df_melted, x="NU_ANO_CENSO", y="Crescimento (%)", hue="Categoria")
plt.xlabel("Ano do Censo")
plt.ylabel("Crescimento (%)")
plt.title("Crescimento Anual de Ingressantes e Concluintes (%)")
plt.axhline(0, color='gray', linestyle='--')  # Add baseline at 0%
plt.legend(title="Categoria")
plt.grid(True)
plt.show()
```
